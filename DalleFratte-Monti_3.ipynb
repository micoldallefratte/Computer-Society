# Descrizione dataframe
Distribuzione degli studenti che si iscrivono ai **Corsi di studio per diploma**, sia in termini di valore assoluto che di percentuale.

I dati si riferiscono alle coorti (gruppo di studenti che iniziano nello stesso anno accademico la propria carriera universitaria) dal 2015/2016 al 2017/2018, e ai corsi di Laurea o Laurea Magistrale a ciclo unico.

Le percentuali sono indicate come valori compresi tra 0 (equivalente a 0%) e 1 (100%).


**corso_descrizione**:	       Descrizione del corso di studi

**studenti**:                  Totale degli studenti della coorte

**p_diploma_professionale**:   % diploma di Istituti Professionali

**p_diploma_tecnico**:         % diploma di Istituti Tecnici

**p_diploma_magistrale**:      % diploma di Istituti Magistrali e Licei Psico - Pedagogici

**p_diploma_liceo**:           % diploma di Licei Scientifici, Classici e Linguistici

**p_diploma_altroitaliano**:   % diploma di altri istituti italiani

**p_diploma_estero**:          % diploma estero

**p_diploma_nondefinito**:     % altro diploma non definito

```python
#merge
coorti= pd.merge(coorti_per_genere, coorti_per_diploma)
coorti
```

```python
#trasformo il nome del corso nel nuovo indice
coorti_= coorti.set_index('corso_descrizione')
```

# ANALISI COORTE 2017-2018
**SELEZIONO CORSI A PREVALENZA MASCHILE NELLA COORTE 2017-2018**

```python
coorti2017_18_M= coorti_.query("coorte == '2017/2018' & prevalenza_genere == 'M'")
coorti2017_18_M
```

```python
max_maschile= pd.DataFrame(coorti2017_18_M, columns=['p_genere_m', 'p_genere_f'])
max_maschile

print("Corsi Max prevalenza maschile:   ",
      max_maschile.loc[(max_maschile['p_genere_f']==0)])     #non ci sono corsi composti completamente da maschi
print("\n")
print("Corsi Max prevalenza maschile:   ",     #trovo la percentuale massima di maschi
      np.max(max_maschile.p_genere_m))
print("\n")
print("Corsi Max prevalenza maschile:   ",
      max_maschile.loc[(max_maschile['p_genere_m']>=0.9)])   #trovo i corsi con la percentuale di maschi piÃ¹ alta
```

- Nella coorte 2017-2018 non ci sono corsi formati da tutti studenti maschi.
- Il massimo di percentuale di maschi Ã¨ del 94% ed Ã¨ raggiunto nel corso di INGEGNERIA MECCANICA, segue poi INGEGNERIA ELETTRONICA PER L'ENERGIA E L'INFORMAZIONE e un altra classe di INGEGNERIA MECCANICA.

**SELEZIONO CORSI A PREVALENZA FEMMINILE NELLA COORTE 2017-2018**

```python
coorti2017_18_F = coorti_.query("coorte == '2017/2018' & prevalenza_genere == 'F'")
coorti2017_18_F
```

```python
max_femminile= pd.DataFrame(coorti2017_18_F, columns=['p_genere_m', 'p_genere_f'])    

print("Max prevalenza femminile:   ",                              #trovo i corsi completamente femminili
      max_femminile.loc[(max_femminile['p_genere_m']==0)])
```

- I corsi completamente formati da studentesse nella coorte 2017-2018 sono: LOGOPEDIA, TECNICHE DI NEUROFISIOPATOLOGIA, CONSERVAZIONE E RESTAURO DEI BENI CULTURALI e OSTETRICIA

# ANALISI GRAFICA

```python
#librerie grafiche
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Analizzo la distribuzione di genere a seconda dei diversi diplomi

```python
#diploma liceo
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_liceo',
            hue="prevalenza_genere", hue_order=['M', 'F'], palette="pastel")  
plt.xlabel("genere")
plt.ylabel("% diploma liceo")
plt.title("Studenti Diplomati al Liceo")
```

```python
#diploma istituto tecnico
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_tecnico', 
            hue="prevalenza_genere", hue_order=['M', 'F'], palette="pastel")   
plt.xlabel("genere")
plt.ylabel("% diploma tecnico")
plt.title("Studenti Diplomati all'Istituto Tecnico")
```

```python
#diploma istituto professionale
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_professionale', 
            hue="prevalenza_genere", hue_order=['M', 'F'], palette="pastel")   
plt.xlabel("genere")
plt.ylabel("% diploma professionale")
plt.title("Studenti Diplomati all'Istituto Professionale")
```

- La percentuale di maschi e femmine diplomati al liceo Ã¨ pressochÃ¨ uguale, ed Ã¨ superiore rispetto agli altri istituti
- Gli studenti universitari diplomati negli istituti tecnici sono di maggioranza maschi
- Gli studenti diplomati negli istituti professionali sono a maggioranza femmine, tuttavia la percentuale Ã¨ molto bassa rispetti ai diplomati nei licei o istituti tecnici

```python
#diploma all'estero
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_estero', 
            hue="prevalenza_genere", hue_order=['M','F'], palette="pastel")   
plt.xlabel("genere")
plt.ylabel("% diploma all'estero")
plt.title("Studenti Diplomati all'Estero")
```

La percentuale di studenti diplomati all'estero Ã¨ molto bassa e la maggioranza tra maschi e femmine Ã¨ irrilevante.
