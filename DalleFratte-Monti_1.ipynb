# Descrizione dataframe
Il dataframe scelto presenta alcuni indicatori che riassumono le caratteristiche dei corsi di studio dell'università di Bologna: internazionalizzazione, regolarità negli studi, soddisfazione degli studenti e ingresso nel mondo del lavoro.


## Periodo di riferimento degli indicatori
Le **percentuali** sono indicate come valori compresi tra 0 (equivalente a 0%) e 1 (100%).

**Studenti internazionali e fuori regione**: il dato è riferito agli studenti immatricolati nell’a.a. 2016/17.

**Laureati in corso**: il dato è riferito agli studenti immatricolati nell'a.a. 2013/14 (lauree triennali), 2014/15 (lauree magistrali), 2011/12 (lauree magistrali a ciclo unico di durata 5 anni), 2010/11 (lauree magistrali a ciclo unico di durata 6 anni).

**Laureati con un'esperienza all'estero**: il dato è riferito ai laureati nell’anno solare 2016.

**Laureati soddisfatti degli studi svolti**: il dato è riferito ai laureati nell’anno solare 2016, indagine AlmaLaurea.

**Laureati che lavorano** e **laureati che non lavorano ma studiano o non cercano**: il dato è riferito ai laureati nell’anno solare 2015 intervistati nel 2016 dopo un anno dal conseguimento del titolo, indagine AlmaLaurea.

```python
#del df di riferimento prendo solo le colonne che intendo analizzare nello specifico
uni= pd.DataFrame(data, columns=['corso_descrizione', 'p_laureatiincorso',
                                        'p_esperienzaestero','p_soddisfatti','p_lavorano','p_studiano'])

uni.rename(columns={"corso_descrizione":"corso"}, inplace=True)

print(uni)
```

**Creo dei sottogruppi del dataframe 'uni' che raccolgano i corsi di laurea in macroaree**

```python
facoltà_economiche= uni.loc[(uni['corso'].str.contains('Economia'))|
                            (uni['corso'].str.contains('marketing'))|
                            (uni['corso'].str.contains('management'))|
                            (uni['corso'].str.contains('aziendale'))|
                            (uni['corso'].str.contains('finanza'))|
                            (uni['corso'].str.contains('amministrazione'))|
                            (uni['corso'].str.contains('Consulente'))|
                            (uni['corso'].str.contains('Giurista'))|                                      
                            (uni['corso'].str.contains('economics'))]
print(facoltà_economiche)
```

```python
facoltà_ingegneria= uni.loc[(uni['corso'].str.contains('Ingegneria'))|
                            (uni['corso'].str.contains('ingegneria'))]

print(facoltà_ingegneria)
```

```python
#facoltà umanistiche 
facoltà_umanistiche= uni.loc[(uni['corso'].str.contains('Filosofia'))|
                            (uni['corso'].str.contains('Lettere'))|
                            (uni['corso'].str.contains('Storia'))|
                            (uni['corso'].str.contains('Psicologia'))|
                            (uni['corso'].str.contains('Educatore'))|
                            (uni['corso'].str.contains('Pedagogia'))|
                            (uni['corso'].str.contains('Scienze della comunicazione'))|
                            (uni['corso'].str.contains('Scienze filosofiche'))|
                            (uni['corso'].str.contains('Scienze storiche'))|
                            (uni['corso'].str.contains('Scienze del comportamento e delle relazioni sociali'))|
                            (uni['corso'].str.contains('Servizio sociale'))|
                            (uni['corso'].str.contains('Sociologia'))|
                            (uni['corso'].str.contains('Scienze della comunicazione'))]
print(facoltà_umanistiche)
```

## ANALISI VARIABILE *PERCENTUALE LAUREATI IN CORSO* 

```python
#statistiche
import numpy as np
print("Media laureati in corso facoltà economiche:   ",
      np.mean(facoltà_economiche.p_laureatiincorso))
print("Media laureati in corso facoltà ingegneria:   ",
      np.mean(facoltà_ingegneria.p_laureatiincorso))
print("Media laureati in corso facoltà umanistiche :   ",
      np.mean(facoltà_umanistiche.p_laureatiincorso))
```

La media percentuale di laureati in corso delle facoltà economiche è circa del **63%**.

La media percentuale di laureati in corso delle facoltà ingegneristiche è del **41%**.

La media percentuale di laureati in corso delle facoltà umanistiche è del **53%**.


Ciò significa che:
1. le facoltà economiche hanno un **22%** in più di laureati in corso rispetto alle facoltà ingegneristiche e un **10%** in più rispetto a quelle umanistiche
2. la maggior parte dei laureati in ingegneria sono fuori corso, infatti solo il **41%** si laurea in tempo
3. le facoltà umanistiche sono seconde rispetto alle facoltà economiche ma superano del **12%** quelle ingegneristiche per percentuale di laureati in corso

*(questa analisi è puramente numerica e non tiene conto di variabili esterne che possono influenzare queste statistiche come: difficoltà dei corsi di laurea, mole di impegno richiesto, studenti lavoratori ecc...)*

Analizzo ora la variabile *percentuale esperienze all'estero* nelle diverse facoltà
