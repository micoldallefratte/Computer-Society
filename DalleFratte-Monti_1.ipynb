# Descrizione dataframe
Il dataframe scelto presenta alcuni indicatori che riassumono le caratteristiche dei corsi di studio dell'universitÃ  di Bologna: internazionalizzazione, regolaritÃ  negli studi, soddisfazione degli studenti e ingresso nel mondo del lavoro.


## Periodo di riferimento degli indicatori
Le **percentuali** sono indicate come valori compresi tra 0 (equivalente a 0%) e 1 (100%).

**Studenti internazionali e fuori regione**: il dato Ã¨ riferito agli studenti immatricolati nellâ€™a.a. 2016/17.

**Laureati in corso**: il dato Ã¨ riferito agli studenti immatricolati nell'a.a. 2013/14 (lauree triennali), 2014/15 (lauree magistrali), 2011/12 (lauree magistrali a ciclo unico di durata 5 anni), 2010/11 (lauree magistrali a ciclo unico di durata 6 anni).

```python
#del df di riferimento prendo solo le colonne che intendo analizzare nello specifico
uni= pd.DataFrame(data, columns=['corso_descrizione', 'p_laureatiincorso',
                                        'p_esperienzaestero','p_soddisfatti','p_lavorano','p_studiano'])

uni.rename(columns={"corso_descrizione":"corso"}, inplace=True)

print(uni)
```

**Creo dei sottogruppi del dataframe 'uni' che raccolgano i corsi di laurea in macroaree**

```python
facoltÃ _economiche= uni.loc[(uni['corso'].str.contains('Economia'))|
                            (uni['corso'].str.contains('marketing'))|
                            (uni['corso'].str.contains('management'))|
                            (uni['corso'].str.contains('aziendale'))|
                            (uni['corso'].str.contains('finanza'))|
                            (uni['corso'].str.contains('amministrazione'))|
                            (uni['corso'].str.contains('Consulente'))|
                            (uni['corso'].str.contains('Giurista'))|                                      
                            (uni['corso'].str.contains('economics'))]
print(facoltÃ _economiche)
```

```python
facoltÃ _ingegneria= uni.loc[(uni['corso'].str.contains('Ingegneria'))|
                            (uni['corso'].str.contains('ingegneria'))]

print(facoltÃ _ingegneria)
```

```python
#facoltÃ  umanistiche 
facoltÃ _umanistiche= uni.loc[(uni['corso'].str.contains('Filosofia'))|
                            (uni['corso'].str.contains('Lettere'))|
                            (uni['corso'].str.contains('Storia'))|
                            (uni['corso'].str.contains('Psicologia'))|
                            (uni['corso'].str.contains('Educatore'))|
                            (uni['corso'].str.contains('Pedagogia'))|
                            (uni['corso'].str.contains('Scienze della comunicazione'))|
                            (uni['corso'].str.contains('Scienze filosofiche'))|
                            (uni['corso'].str.contains('Scienze storiche'))|
                            (uni['corso'].str.contains('Scienze del comportamento e delle relazioni sociali'))|
                            (uni['corso'].str.contains('Servizio sociale'))|
                            (uni['corso'].str.contains('Sociologia'))|
                            (uni['corso'].str.contains('Scienze della comunicazione'))]
print(facoltÃ _umanistiche)
```

## ANALISI VARIABILE *PERCENTUALE LAUREATI IN CORSO* 

```python
#statistiche
import numpy as np
print("Media laureati in corso facoltÃ  economiche:   ",
      np.mean(facoltÃ _economiche.p_laureatiincorso))
print("Media laureati in corso facoltÃ  ingegneria:   ",
      np.mean(facoltÃ _ingegneria.p_laureatiincorso))
print("Media laureati in corso facoltÃ  umanistiche :   ",
      np.mean(facoltÃ _umanistiche.p_laureatiincorso))
```

La media percentuale di laureati in corso delle facoltÃ  economiche Ã¨ circa del **63%**.

La media percentuale di laureati in corso delle facoltÃ  ingegneristiche Ã¨ del **41%**.

La media percentuale di laureati in corso delle facoltÃ  umanistiche Ã¨ del **53%**.


CiÃ² significa che:
1. le facoltÃ  economiche hanno un **22%** in piÃ¹ di laureati in corso rispetto alle facoltÃ  ingegneristiche e un **10%** in piÃ¹ rispetto a quelle umanistiche
2. la maggior parte dei laureati in ingegneria sono fuori corso, infatti solo il **41%** si laurea in tempo
3. le facoltÃ  umanistiche sono seconde rispetto alle facoltÃ  economiche ma superano del **12%** quelle ingegneristiche per percentuale di laureati in corso

*(questa analisi Ã¨ puramente numerica e non tiene conto di variabili esterne che possono influenzare queste statistiche come: difficoltÃ  dei corsi di laurea, mole di impegno richiesto, studenti lavoratori ecc...)*
