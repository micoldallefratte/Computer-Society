### Importo il primo dataframe

```python
import pandas as pd
coorti_per_genere = pd.read_csv(r"C:\Users\admin\Desktop\COMPUTER&SOCIETY\punti aggiuntivi\coorte-per-genere_all.csv",
                                usecols=['coorte','corso_descrizione','studenti', 'p_genere_m','p_genere_f'])
coorti_per_genere
```

```python
#aggiungere nuova colonna 'prevalenza_genere' che mi mostri in modo più intuitivo la maggioranza di genere di ciascun corso

import numpy as np
coorti_per_genere['prevalenza_genere']= np.where(coorti_per_genere['p_genere_m']>coorti_per_genere['p_genere_f'], 'M', 'F' )
coorti_per_genere
```

# Descrizione dataframe
Distribuzione degli studenti che si iscrivono ai **Corsi di studio per genere**, sia in termini di valore assoluto che di percentuale.

I dati si riferiscono alle coorti (gruppo di studenti che iniziano nello stesso anno accademico la propria carriera universitaria) dal 2015/2016 al 2017/2018.

Le percentuali sono indicate come valori compresi tra 0 (equivalente a 0%) e 1 (100%).


**corso_descrizione**:	Descrizione del corso di studi

**studenti**:           Totale degli studenti della coorte

**p_genere_m**:         % di studenti

**p_genere_f**:	        % di studentesse

### Importo secondo dataframe

```python
coorti_per_diploma = pd.read_csv(r"C:\Users\admin\Desktop\COMPUTER&SOCIETY\punti aggiuntivi\coorte-per-diploma_all.csv",
                                 usecols=['coorte','corso_descrizione','studenti', 'p_diploma_professionale','p_diploma_tecnico',
                                'p_diploma_magistrale','p_diploma_liceo','p_diploma_altroitaliano','p_diploma_estero','p_diploma_nondefinito'])
coorti_per_diploma
```

# Descrizione dataframe
Distribuzione degli studenti che si iscrivono ai **Corsi di studio per diploma**, sia in termini di valore assoluto che di percentuale.

I dati si riferiscono alle coorti (gruppo di studenti che iniziano nello stesso anno accademico la propria carriera universitaria) dal 2015/2016 al 2017/2018, e ai corsi di Laurea o Laurea Magistrale a ciclo unico.

Le percentuali sono indicate come valori compresi tra 0 (equivalente a 0%) e 1 (100%).


**corso_descrizione**:	       Descrizione del corso di studi

**studenti**:                  Totale degli studenti della coorte

**p_diploma_professionale**:   % diploma di Istituti Professionali

**p_diploma_tecnico**:         % diploma di Istituti Tecnici

**p_diploma_magistrale**:      % diploma di Istituti Magistrali e Licei Psico - Pedagogici

**p_diploma_liceo**:           % diploma di Licei Scientifici, Classici e Linguistici

**p_diploma_altroitaliano**:   % diploma di altri istituti italiani

**p_diploma_estero**:          % diploma estero

**p_diploma_nondefinito**:     % altro diploma non definito

```python
#merge
coorti= pd.merge(coorti_per_genere, coorti_per_diploma)
coorti
```

## Analisi missing data

```python
coorti.isnull().any() #non ci sono missing data
```

```python
#trasformo il nome del corso nel nuovo indice
coorti_= coorti.set_index('corso_descrizione')
```

# ANALISI COORTE 2017-2018
**SELEZIONO CORSI A PREVALENZA MASCHILE NELLA COORTE 2017-2018**

```python
coorti2017_18_M= coorti_.query("coorte == '2017/2018' & prevalenza_genere == 'M'")
coorti2017_18_M
```

```python
max_maschile= pd.DataFrame(coorti2017_18_M, columns=['p_genere_m', 'p_genere_f'])
max_maschile

print("Corsi Max prevalenza maschile:   ",
      max_maschile.loc[(max_maschile['p_genere_f']==0)])     #non ci sono corsi composti completamente da maschi
print("\n")
print("Corsi Max prevalenza maschile:   ",     #trovo la percentuale massima di maschi
      np.max(max_maschile.p_genere_m))
print("\n")
print("Corsi Max prevalenza maschile:   ",
      max_maschile.loc[(max_maschile['p_genere_m']>=0.9)])   #trovo i corsi con la percentuale di maschi più alta
```

- Nella coorte 2017-2018 non ci sono corsi formati da tutti studenti maschi.
- Il massimo di percentuale di maschi è del 94% ed è raggiunto nel corso di INGEGNERIA MECCANICA, segue poi INGEGNERIA ELETTRONICA PER L'ENERGIA E L'INFORMAZIONE e un altra classe di INGEGNERIA MECCANICA.

**SELEZIONO CORSI A PREVALENZA FEMMINILE NELLA COORTE 2017-2018**

```python
coorti2017_18_F = coorti_.query("coorte == '2017/2018' & prevalenza_genere == 'F'")
coorti2017_18_F
```

```python
max_femminile= pd.DataFrame(coorti2017_18_F, columns=['p_genere_m', 'p_genere_f'])    

print("Max prevalenza femminile:   ",                              #trovo i corsi completamente femminili
      max_femminile.loc[(max_femminile['p_genere_m']==0)])
```

- I corsi completamente formati da studentesse nella coorte 2017-2018 sono: LOGOPEDIA, TECNICHE DI NEUROFISIOPATOLOGIA, CONSERVAZIONE E RESTAURO DEI BENI CULTURALI e OSTETRICIA

# ANALISI GENERALE DALLA CORTE 2015 ALLA 2018

## Analizzo la distribuzione della provenienza in base al genere dal 2015 al 2018
**Utilizzo la funzione *groupby***

```python
coorti_genere = coorti_.groupby(['prevalenza_genere'])
coorti_genere.sum()
```

**Commento**: 
- la maggioranza delle studentesse femmine proviene da un liceo 
- la maggioranza di studenti maschi proviene da un liceo

Si può quindi affermare che, indipendentemente dal genere, la maggioranza degli studenti iscritti all'università provengono da un liceo, seguono poi gli istituti tecnici ed infine una quantità esigua proviene da istituti professionali.

## Analizzo quali sono i corsi con una prevalenza di genere femminile sopra il 60% 

```python
coorti_F=coorti_.query("p_genere_f>0.60")
coorti_F
```

La tabella ottenuta mi permette di osservare tutti i corsi che hanno una prevalenza di genere femminile sopra il 60%, quindi abbastanza consistente.
Posso notare che i corsi in cui prevale il genere femminile sono principalmente di tipo **umanistico-sociale** e di **professioni sanitarie**.

## Analizzo quali sono i corsi con una prevalenza di genere maschile sopra il 60%

```python
coorti_M=coorti_.query("p_genere_m>0.60")
coorti_M
```

La tabella ottenuta mi permette di osservare quali sono i corsi il cui il genere maschile è sopra il 60%.
Analizzado i dati ottenuti dalla possiamo osservare che i corsi in cui prevale il genere maschile sono di tipo **economico ed ingegneristico**.

# ANALISI GRAFICA

```python
#librerie grafiche
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

## Analizzo la distribuzione di genere a seconda dei diversi diplomi

```python
#diploma liceo
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_liceo')  
plt.xlabel("genere")
plt.ylabel("% diploma liceo")
plt.title("Studenti Diplomati al Liceo")
```

```python
#diploma istituto tecnico
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_tecnico')   
plt.xlabel("genere")
plt.ylabel("% diploma tecnico")
plt.title("Studenti Diplomati all'Istituto Tecnico")
```

```python
#diploma istituto professionale
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_professionale')   
plt.xlabel("genere")
plt.ylabel("% diploma professionale")
plt.title("Studenti Diplomati all'Istituto Professionale")
```

- La percentuale di maschi e femmine diplomati al liceo è pressochè uguale, ed è superiore rispetto agli altri istituti
- Gli studenti universitari diplomati negli istituti tecnici sono di maggioranza maschi
- Gli studenti diplomati negli istituti professionali sono a maggioranza femmine, tuttavia la percentuale è molto bassa rispetti ai diplomati nei licei o istituti tecnici

```python
#diploma all'estero
sns.barplot(data=coorti, x='prevalenza_genere', y='p_diploma_estero')   
plt.xlabel("genere")
plt.ylabel("% diploma all'estero")
plt.title("Studenti Diplomati all'Estero")
```

La percentuale di studenti diplomati all'estero è molto bassa e la maggioranza tra maschi e femmine è irrilevante.
